```python
import requests
import pandas as pd
import json
import os
```

```python
def calculator(expression):
    try:
        allowed_chars = "0123456789+-*/(). "
        if all(char in allowed_chars for char in expression):
            return eval(expression)
        else:
            return "Invalid characters in expression"
    except Exception as e:
        return f"Calculation error: {str(e)}"
```

```python
def web_search(query):
    try:
        url = f"https://api.duckduckgo.com/?q={query}&format=json"
        response = requests.get(url).json()

        if response.get("Abstract"):
            return response["Abstract"]
        elif response.get("RelatedTopics"):
            return response["RelatedTopics"][0].get("Text", "No results found")
        else:
            return "No results found"

    except Exception as e:
        return f"Search error: {str(e)}"
```

```python
def file_reader(filename):
    try:
        if not os.path.exists(filename):
            return "File not found"

        # txt
        if filename.endswith(".txt"):
            with open(filename, "r") as f:
                return f.read()

        # json
        elif filename.endswith(".json"):
            with open(filename, "r") as f:
                return json.dumps(json.load(f), indent=2)

        # csv
        elif filename.endswith(".csv"):
            df = pd.read_csv(filename)
            return df.head().to_string()

        else:
            return "Unsupported file type"

    except Exception as e:
        return f"File read error: {str(e)}"
```

```python
def agent(user_input):

    user_input = user_input.strip()

    # read file
    if user_input.lower().startswith("read"):
        filename = user_input[5:].strip()
        return file_reader(filename)

    # Web search
    elif user_input.lower().startswith("search"):
        query = user_input[7:].strip()
        return web_search(query)

    # calc
    elif any(char.isdigit() for char in user_input):
        return calculator(user_input)

    else:
        return "Sorry, I couldn't understand your request."
```

```python
if __name__ == "__main__":
    print("Tool Agent Ready! (type 'exit' to quit)")

    while True:
        user_input = input("\nEnter your query: ")

        if user_input.lower() == "exit":
            break

        result = agent(user_input)
        print("RESULT\n", result)
```
