```python
def summarize(text):
    try:
        response = model.generate_content(
            f"Summarize the following text clearly:\n{text}"
        )
        return response.text
    except Exception as e:
        return f"Error: {str(e)}"
```

```python
def read_file(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            return f.read()
    except:
        return "Error: File not found"
```

```python
def generate_email(summary):
    return f"""
Hi,

Here is the summary of the document:

{summary}

Best regards,
AI Agent
"""
```

```python
def send_email(to_email, subject, body):
    sender_email = "mdnoufal.official@gmail.com"
    app_password = "dwnq dexf bxjp qbwu"  # Use App Password

    msg = MIMEText(body)
    msg["Subject"] = subject
    msg["From"] = sender_email
    msg["To"] = to_email

    try:
        server = smtplib.SMTP("smtp.gmail.com", 587)
        server.starttls()
        server.login(sender_email, app_password)
        server.send_message(msg)
        server.quit()

        return " Email sent successfully!"
    except Exception as e:
        return f" Email error: {str(e)}"
```

```python
def task_agent(user_input):

    if "summarize" in user_input.lower() and "email" in user_input.lower():

        # Step 1: Get file
        filename = input("Enter file name: ")
        content = read_file(filename)

        if content.startswith("Error"):
            return content

        # Step 2: Summarize
        print(" Summarizing document...")
        summary = summarize(content)

        if summary.startswith("Error"):
            return summary

        # Step 3: Generate email
        email_body = generate_email(summary)

        # Step 4: Send email
        to_email = input("Enter recipient email: ")
        print(" Sending email...")

        result = send_email(to_email, "Document Summary", email_body)

        return result

    else:
        return "Task not recognized"
```
