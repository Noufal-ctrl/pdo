```python
import os
import google.generativeai as genai

os.environ["GEMINI_API_KEY"] = "AIzaSyDls6axVUd0tOnKeAjcP4mck2eCBEhtW5cVUDwaji"

genai.configure(api_key=os.environ["GEMINI_API_KEY"])


model = genai.GenerativeModel("models/gemini-2.5-flash")


def ask_llm(prompt):
    response = model.generate_content(prompt)
    return response.text


TEMPLATES = {
    "summarize": "Summarize in 3 lines:",
    "explain":   "Explain simply:",
    "email":     "Write a professional email:"
}


def run_prompt(task, text):
    if task not in TEMPLATES:
        return "Invalid task"
    prompt = f"{TEMPLATES[task]}\n{text}"
    return ask_llm(prompt)


def summarize(text):      return run_prompt("summarize", text)
def explain(text):        return run_prompt("explain",   text)
def generate_email(text): return run_prompt("email",     text)


# 🔹 Test
if __name__ == "__main__":
    print("---- Summarize ----")
    print(summarize("Machine learning is a subset of AI that learns from data."))

    print("\n---- Explain ----")
    print(explain("Caching improves performance by storing frequently accessed data."))

    print("\n---- Email ----")
    print(generate_email("Request leave for 2 days due to personal reasons"))

    print("\n---- Reusable Engine ----")
    print(run_prompt("summarize", "Data engineering involves building pipelines."))
```
