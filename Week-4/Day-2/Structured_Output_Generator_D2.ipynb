````python
from mistralai.client import Mistral
import json
import os


api_key = "sTtbKxZpJ6xkk1TYPnjzwdjn4njDN2pC2I4Qej8fxPzrC"
client = Mistral(api_key=api_key)


def extract_flight_details(user_input):
    prompt = f"""
    Extract structured data from the user input.

    Return ONLY valid JSON. Do NOT include explanation.

    Format:
    {{
      "source": "<city>",
      "destination": "<city>",
      "date": "<date>"
    }}

    Rules:
    - If any field is missing, return null
    - Output must be valid JSON only

    User Input: "{user_input}"
    """

    response = client.chat.complete(
        model="mistral-small-latest",
        messages=[
            {"role": "user", "content": prompt}
        ]
    )

    text = response.choices[0].message.content.strip()


    if text.startswith('```json') and text.endswith('```'):
        text = text[len('```json'):-len('```')].strip()

    try:
        data = json.loads(text)
        return data
    except json.JSONDecodeError:
        return {
            "error": "Invalid JSON output",
            "raw_output": text
        }


if __name__ == "__main__":
    result = extract_flight_details("Book flight from Chennai to Delhi tomorrow")
    print(result)
````
