```python
import pandas as pd
```

```python
df = pd.read_csv("user_transactions.csv")
```

```python
user_total = df.groupby("user")["amount"].sum().reset_index()

print("Total Amount per User:")
print(user_total)
```

```python
top_spender = user_total.loc[user_total["amount"].idxmax()]

print("\nTop Spender:")
print(top_spender)
```

```python
print("--- Summary Report ---")
print(f"Total Users: {df['user'].nunique()}")
print(f"Total Transactions: {len(df)}")
print(f"Total Amount Spent: {df['amount'].sum()}")
print(f"Top Spender: {top_spender['user']} with {top_spender['amount']}")
```
