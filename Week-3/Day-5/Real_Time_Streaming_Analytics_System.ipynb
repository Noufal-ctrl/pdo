```python
import time
import threading
import random
from collections import defaultdict
```

```python
click_data = ["user1", "user2", "user3", "user1", "user2"]
```

```python
click_count=defaultdict(int)
```

```python
def stream_data():
  while True:
    user = random.choice(click_data)
    click_count[user]+=1
    print(f"[STREAM] {user} clicked")
    time.sleep(1)
```

```python
def analytics():
    while True:
        time.sleep(5)
        print("[ANALYTICS   --->  Last 5 sec]")
        for user, count in click_count.items():
            print(f"{user}: {count} clicks")
        print("----------------------")
```

```python
t1 = threading.Thread(target=stream_data)
t2 = threading.Thread(target=analytics)
t1.start()
t2.start()
t1.join()
t2.join()
```
