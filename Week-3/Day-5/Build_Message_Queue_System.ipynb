```python
import time
import threading
```

```python
import random
```

```python
import queue
```

```python
msg_queue = queue.Queue()
```

```python
def producer():
  messages = ["order_created", "payment_done", "order_delivered"]
  for i in range(10):
    msg = random.choice(messages)
    msg_queue.put(msg)
    print(f"[PRODUCER] Sent : {msg}")
    time.sleep(1)
  msg_queue.put(None)
  print("[PRODUCER] Finished sending messages.")
```

```python
def consumer():
  while True:
    if not msg_queue.empty():
      msg = msg_queue.get()
      if msg is None:
        print("[CONSUMER] Received stop signal. Exiting.")
        break
      print(f"[CONSUMER] Processing : {msg}")
      time.sleep(2)
    else:
      time.sleep(1)
```

```python
t1 = threading.Thread(target=producer)
t2 = threading.Thread(target=consumer)
t1.start()
t2.start()
t1.join()
t2.join()
```

```python
t1.stop()
```
