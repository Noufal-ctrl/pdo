```python
import logging
```

```python
import time
import sys
```

```python
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(levelname)s - %(message)s',
    stream=sys.stdout,
    force=True
    )
```

```python
def extract():
    logging.info("Extracting data")
    data = [
        {"user": "A", "amount": 100},
        {"user": "B", "amount": None},
        {"user": "A", "amount": 300}
    ]
    return data
```

```python
def clean(data):
  logging.info("Cleaning Data")
  cleaned = [d for d in data if d['amount'] is not None]
  return cleaned
```

```python
def aggregate(data):
    logging.info("Aggregating data")
    result = {}
    for d in data:
        user = d["user"]
        result[user] = result.get(user, 0) + d["amount"]
    return result
```

```python
def run_pipeline():
    logging.info("Pipeline Started")

    data = extract()
    cleaned = clean(data)
    aggregated = aggregate(cleaned)

    logging.info(f"Final Output: {aggregated}")
    logging.info("Pipeline Completed")
```

```python
run_pipeline()
```
